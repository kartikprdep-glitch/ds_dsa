```python
class Costomer:
    def __init__(self,id,name,address):
      self.id=id
      self.name=name
      self.address=address
    def cust_info(self):
      print(f"id{self.id} name{self.name} address{self.address}")
```

```python
c=Costomer(100,"joe","banglore")
```

```python
c.cust_info
```

```python
class Cart:
    def __init__(self,name,code,price,qty,):
      self.name=name
      self.code=code
      self.price=price
      self.qty=qty
    def cart_info(self):
        total= self.price*self.qty
        print(f"name{self.name} code{self.code} price:{self.price} qty:{self.qty} total:{total}")
```

```python
cart=Cart("laptop",1000,36000,2)
```

```python
cart.cart_info()
```
