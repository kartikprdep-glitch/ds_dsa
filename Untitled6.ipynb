```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
```

```python
n1=Node(5)
n2=Node(7)
n3=Node(10)
n4=Node(12)
n5=Node(19)
```

```python
n1.next=n2
n2.next=n3
n3.next=n4
n4.next=n5
```

```python
cur=n1
odd_sum=0
while cur:
    if cur.data %2==1:
        odd_sum+=cur.data
    cur=cur.next
print(odd_sum)
```
