```python
class Order:
    def __init__(self,id,date,price,qty):
        self.id=id
        self.date=date
        self.price=price
        self.qty=qty
    def order_info(self):
        print(f"Orderid:{self.id} OrderDate:{self.date} Orderprice:{self.price} Orderqty:{self.qty}")
```

```python
o1=Order(100,"10-10-2026",2200,2)
o2=Order(101,"15-7-2026",25000,3)
o3=Order(300,"04-6-2025",10000,4)
o4=Order(400,"04-8-2026",25000,5)
```

```python
o1.order_info()
```

```python
o2.order_info()
```

```python
o3.order_info()
```

```python
o4.order_info()
```

```python
from collections import deque
```

```python
order_queue=deque()
```

```python
order_queue.append(o1)
order_queue.append(o2)
order_queue.append(o3)
order_queue.append(o4)
```

```python
while order_queue:
    cur=order_queue.popleft()
    print("processing order")
    cur.order_info()
```
