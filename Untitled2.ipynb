```python
#create a linked list and,1)find the sum of all nodes.2)find the max node.3)reverse the nodes.
```

```python
n1=Node(10)
n2=Node(20)
n3=Node(30)
n4=Node(40)
```

```python
n1.next=n2
n2.next=n3
n3.next=n4
```

```python
s=0
cur=n1
while cur:
    s+=cur.data
    cur=cur.next
print(s)
```

```python
max=0
cur=n1
while cur:
    if cur.data>max:
        max=cur.data
        cur=cur.next
print(max)
```

```python
rev=[]
cur=n1
while cur:
    rev.insert(0,cur.data)
    cur=cur.next
print(rev)    
```

```python
class Student:
    def __init__(self,id,name,result):
        self.id=id
        self.name=name
        self.result=result
    def stu_info(self):
        print(f"StudentId:{self.id} StudentName:{self.name} StudentResult:{self.result} " )    
```
