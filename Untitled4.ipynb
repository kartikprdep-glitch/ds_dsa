```python
class Node :
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
```

```python
n1=Node(30)
n2=Node(40)
n3=Node(50)
n4=Node(60)
n5=Node(35)
```

```python
n1.next=n5
n5.next=n2
n2.prev=n5
n2.next=n3
n3.next=n4
```

```python
cur=n1
while cur:
    print(cur.data)
    cur=cur.next
```
